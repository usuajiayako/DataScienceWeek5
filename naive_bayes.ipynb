```python
import numpy as np
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
data = pd.read_csv('food_data.csv')  # load our data
data
```

```python
data.index = data.iloc[:,0] # Name rows as food for reference, making a column to the index
data_to_use = data.iloc[:,3:16]
data_to_target = data.iloc[:,16]

data_to_target
```

```python
# 70% training, 30% test
X_train, X_test, y_train, y_test = train_test_split(data_to_use, data_to_target, test_size=0.3, random_state=21)
```

```python
# define a model (Gaussian model)
gaussian_model = GaussianNB()
```

```python
# Training the model
gaussian_model.fit(X_train, y_train)
```

```python
# Predict the response
prediction = gaussian_model.predict(X_test)
```

```python
# making comparison between actual and prediction
result = {'food_type':y_test, 'pred':prediction}
```

```python
# converting the result to dataframe
result_df = pd.DataFrame(result)
```

```python
# Checking NB accuracy
print("Accuracy:", metrics.accuracy_score(y_test, prediction)) 
```
